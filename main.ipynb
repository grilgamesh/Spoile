```python
#import dependencies
from splinter import Browser
from bs4 import BeautifulSoup as bs
import time
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
import re
import fnmatch
import random
```

```python
# Set up Splinter
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
# initial experiment: just stick to page 1. We can stick this in a loop and increase data arbitrarily; 
#i was still recogising films on page 50...
page_number = 1
url = f"https://www.imdb.com/search/keyword/?ref_=kw_nxt&mode=detail&page={page_number}&sort=num_votes,desc&title_type=movie"
film_names = []
film_ids = []
```

```python
def populate_films(url):
    browser.visit(url)
    soup = bs(browser.html, 'html.parser')
    href_list = soup.find_all('img', class_='loadlate')
    for item in href_list:
        film_names.append(item['alt'])
        film_ids.append(item['data-tconst'])
```

```python
populate_films(url)
print(film_names)
```

```python
print(film_ids)
```

```python
def get_keywords(film_id):
    keyURL = f"https://www.imdb.com/title/{film_id}/keywords"
    browser.visit(keyURL)
    soup = bs(browser.html, 'html.parser')
    key_list = soup.find_all('div', class_='sodatext')
    
    keys = []
    for div in key_list:
        keys.append(div.find('a').contents[0])
    return keys
    
```

```python
film_dict = {}
for i in range(0, len(film_names)):
    new_film = {'id': film_ids[i], 'tags': get_keywords(film_ids[i])}
    film_dict.update({film_names[i]: new_film})
```

```python
answer = random.choice(film_names)
print("i have chose a random film")
print("work out the film by guessing. every guess will reveal more tags")
guess = ""
tag_list = []
```

```python
while guess != answer:
    guess = input("guess a film!")
    if guess == answer:
        print("you're right! well done")
    else:
        common_tags = set(film_dict[guess]['tags']).intersection(film_dict[answer]['tags'])
        print(f"no, but you have revealed the following tags:")
        print(common_tags.difference(tag_list))
        tag_list = set(common_tags.union(tag_list))
```
